## <font color="orange">Outlier Detection</font>
- Calculate Q1 ( the first Quarter)
- Calculate Q3 ( the third Quartile)
- Find IQR = (Q3 - Q1)
- Find the lower Range = Q1 -(1.5 * IQR)
- Find the upper Range = Q3 + (1.5 * IQR)

```python
def outlier_detection(datacolumn):
    sorted(datacolumn)
    Q1,Q3 = np.percentile(datacolumn , [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    
    return lower_range,upper_range
```

```python
lowerbound,upperbound = outlier_detection(input_df_raw['sales'])

outliers = input_df_raw[(input_df_raw['sales'] < lowerbound) | (input_df_raw['sales'] > upperbound)]
outliers
```

## <font color="orange">Outlier Treatment</font>
We have identified the Outliers in the above give indexes <br>
Since we can remove any data point, as it will create an absentia in the time-series, we will impute the outliers <br>
Our choice of imputation will be <font color="green">_knn-Imputer_</font> as it assigns nulls/nan with the closest knn <br>

```python
# making outliers as NaN so that they can be treated by KNN Imputer
input_df = input_df_raw.copy()
input_df.loc[outliers.index, 'sales']=np.nan
input_df
```

```python
from sklearn.impute import SimpleImputer

impNumeric = SimpleImputer(missing_values=np.nan, strategy='mean')
impCategorical = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

# Fitting the data to the imputer object 
impNumeric = impNumeric.fit(input_df[['sales']])
  
# Imputing the data      
input_df['sales'] = impNumeric.transform(input_df[['sales']])
input_df
```

# <font color="gold"><ins><b>Step4-5: Train-Test Split and Scale Data</b></ins></font>

```python
train, test = df.iloc[:-9], df.iloc[-9:]
print(f"Length of Dataset: {len(df)} weeks\nLength of Train dataset: {len(train)} weeks")
print(f"Length of Test dataset: {len(test)} weeks")
```

```python
from sklearn.preprocessing import MinMaxScaler
import warnings

scaler = MinMaxScaler()

warnings.filterwarnings("ignore")
# Fit only to train dataset, as Test Dataset's scale is unknown
scaler.fit(train)

scaled_train_data = scaler.transform(train)
scaled_test_data = scaler.transform(test)
```

# <font color="gold"><b>Step6: Feeding batches of data to RNN via <ins>Time Series Generator</ins></b></font>

This class takes in a sequence of data-points gathered at
equal intervals, along with time series parameters such as
stride, length of history, etc., to produce batches for
training/validation.


```python
# Step6:
import tensorflow as tf
from tensorflow import keras

# For feeding batches of data to RNN
from keras.preprocessing.sequence import TimeseriesGenerator

# usually equal to the seasonal period cycle, i.e. 7 for day level data, 12 for month level data, etc
n_input = 52
n_features = 1
```

```python
# since source of data and traget are same, pass scaled train dataset twice
train_generator = TimeseriesGenerator(scaled_train_data, scaled_train_data, length=n_input, batch_size=1)
```

```python
print("Len of train_dataset = len of generator + len of n_input")
print(f" {len(scaled_train_data)} = {len(train_generator)} + {n_input}")
```

___
What _TimeseriesGenerator_ does for us is tranform the sequence <br>
> [t1,t2,t3,t4,t5,t6] into <br>
> [t1,t2,t3,t4,t5] -> [t6] <br>
i.e. it takes a sequence of data, transforms it to the above format, i.e. the input format of RNN <br>
and LHS is features, RHS is label for the RNN for each batch <br>
___

# <font color="gold"><ins><b>Step7: Building the LSTM Model</ins></b></font>

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM 

model = Sequential()
model.add((LSTM(150, activation='relu', input_shape=(n_input, n_features))))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
model.fit_generator(train_generator, epochs=400, verbose=2)
epochs_passed_till_now=[400]
```

# <font color="gold"><ins><b>Step10: Predicting using RNN Model</ins></b></font>

```python
# Holds the predections
test_predictions = []

# last n_input points from the training dataset
first_eval_batch = scaled_train_data[-n_input:]
# Reshaping it to the format RNN expects
current_batch = first_eval_batch.reshape((1, n_input, n_features))

# How far to predict?
for i in range(len(test)):
    
    # get prediction 1 time stamp ahead ([0] is for grabbing just the number instead of [array])
    current_pred = model.predict(current_batch)[0]
    
    # store prediction
    test_predictions.append(current_pred) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
    
```

```python
true_predictions_eval_unord = scaler.inverse_transform(test_predictions).tolist()
print(true_predictions_eval_unord)
```

```python
true_predictions_eval = []
for i in range(len(true_predictions_eval_unord)):
    true_predictions_eval.append(true_predictions_eval_unord[i][0])

# print(true_predictions_eval)
test_eval = test.copy()
test_eval['pred'] = true_predictions_eval
test_eval['mape'] = ((test['sales']-test_eval['pred'])/test['sales'])*100
test_eval
```

```python
final_mape = np.mean(test_eval["mape"])
print(f"Final Mape Value: {final_mape}")
```

# <font color="gold"><ins><b>Step10.a) Predicting Future Values using RNN Model</b></ins></font>

```python
# Holds the predections
test_predictions = []
# How many weeks to forecast in future: 3 months, i.e. 12 weeks
additional_weeks_to_fcst=12


# last n_input points from the training dataset
first_eval_batch = scaled_train_data[-n_input:]
# Reshaping it to the format RNN expects
current_batch = first_eval_batch.reshape((1, n_input, n_features))

# How far to predict?
for i in range(len(test)+additional_weeks_to_fcst):
    
    # get prediction 1 time stamp ahead ([0] is for grabbing just the number instead of [array])
    current_pred = model.predict(current_batch)[0]
    
    # store prediction
    test_predictions.append(current_pred) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```
