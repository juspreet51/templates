```python
import pandas as pd
import numpy as np
%matplotlib inline

# Load specific forecasting tools
from statsmodels.tsa.statespace.sarimax import SARIMAX

from statsmodels.graphics.tsaplots import plot_acf,plot_pacf # for determining (p,q) orders
from statsmodels.tsa.seasonal import seasonal_decompose      # for ETS Plots
from pmdarima import auto_arima                              # for determining ARIMA orders

# Ignore harmless warnings
import warnings
warnings.filterwarnings("ignore")

# Load dataset
df = pd.read_csv('../Data/RestaurantVisitors.csv',index_col='date',parse_dates=True)
df.index.freq = 'D'
df.head()
```

## Test for stationarity

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
adf_test(df1['total'])
```

# Adding Exogeneous Variable
### Determining the order of p,d,q and P,D,Q

```python
auto_arima(df1['total'], seasonal=True, m=7).summary()
```

## Train Test Split

```python
# Set four weeks for testing
train, test = df1.iloc[:436], df1.iloc[436:];
print(f'DF:{len(df1)}, Train:{len(train)}, Test:{len(test)}')
```

## Training the Model With <font color='red'>Exogenous Varibale </font>
#### NOTE: To avoid a ValueError: non-invertible starting MA parameters found we're going to set enforce_invertibility to False.

```python
# SARIMAX(1, 0, 0)x(2, 0, 0, 7)
model = SARIMAX(train['total'],exog=train[['holiday']], order=(1,0,0),seasonal_order=(2,0,0,7),enforce_invertibility=False)
result = model.fit()
```

## Predecting From the Model With <font color='red'>Exogenous Varibale </font>


```python
start = len(train); end = len(train)+len(test)-1;
predictions = result.predict(start=start, end=end,exog=test[['holiday']]).rename("SARIMAX(1, 0, 0)x(2, 0, 0, 7) with Exog")
```

```python
ax = test['total'].plot(legend=True);
predictions.plot(legend=True)

for day in test.query('holiday == 1').index:
    ax.axvline(x=day, color='black', alpha=0.9);
```

# Evaluatoin Metrics

```python
from sklearn.metrics import mean_squared_error, r2_score
from statsmodels.tools.eval_measures import rmse

rmse_error = rmse(test['total'],predictions)
r2_score_error = r2_score(test['total'],predictions)

print("Test mean:",test['total'].mean())
print(f'RMSE:{rmse_error} and R2 Sq:{r2_score_error}')
```

# Forecasting the future values

```python
model = SARIMAX(df1['total'],exog=df1['holiday'],order=(1,0,0),seasonal_order=(2,0,0,7),enforce_invertibility=False)
results = model.fit()
exog_forecast = df[478:][['holiday']]

start = len(df1)
fcast = results.predict(start=start, end=start+38, exog=exog_forecast).rename('SARIMAX(1,0,0)(2,0,0,7) Forecast')
```

```python
ax = test['total'].plot(legend=True,color='teal')
fcast.plot(legend=True,color="red");

for day in test.query('holiday == 1').index:
    ax.axvline(x=day, color='black', alpha=0.9);
```
